Логическая операция в программировании - это операция над выражениями логического (булевского) типа, соответствующая некоторой операции над высказываниями в алгебре логики. Как и высказывания, логические выражения могут принимать одно из двух "булевых" значений — «истинно» или «ложно».

логические операции.png

операторы логич.png

Исследования в алгебре логики тесно связаны с изучением высказываний. Высказывание — это языковое образование, в отношении которого имеет смысл говорить о его истинности или ложности (Аристотель).

Простым высказыванием называют повествовательное предложение, относительно которого имеет смысл говорить, истинно оно или ложно.

Считается, что каждое высказывание либо истинно, либо ложно и ни одно высказывание не может быть одновременно истинным и ложным.

Примеры высказываний:

**Москва – столица России.**

**Число 27 является простым.**

**Волга впадает в Каспийское море.**

Следующие предложения высказываниями не являются:

*Давай пойдем гулять.*

*Который час?*



**Итак, отличительным признаком высказывания является свойство быть истинным или ложным, последние два предложения этим свойством не обладают.**

С помощью высказываний устанавливаются свойства, взаимосвязи между объектами. 

Высказывание истинно, если оно адекватно отображает эту связь, в противном случае оно ложно.

Примеры высказываний:

Сегодня светит солнце.
Трава растет.

Каждое из этих высказываний несет значение «истина» или «ложь».

В математической логике не рассматривается конкретное содержание высказывания, важно только, истинно оно или ложно. Поэтому высказывание можно представить некоторой переменной величиной, значением которой может быть только 0 (False) или 1 (True). 

Если высказывание истинно, то его значение равно 1 (True), если ложно - 0 (False).

ДЛЯ ОСМЫСЛЕНИЯ
Однако определение истинности высказывания далеко не простой вопрос. Например, высказывание «Число 1 +22 = 4294 967297 — простое», принадлежащее Ферма (1601-1665), долгое время считалось истинным (True), пока в 1732 году Эйлер (1707-1783) не доказал, что оно ложно (False). В целом, обоснование истинности или ложности простых высказываний решается вне алгебры логики. Например, истинность или ложность высказывания «Сумма углов треугольника равна 180°» устанавливается геометрией, причем в геометрии Евклида это высказывание является истинным (True), а в геометрии Лобачевского — ложным (False).

Не путать блок программы и блок-схему.
Блок-схема - это исключительно схема (фигуры и стрелки), которая служит для описания алгоритма.

А блок программы - это определенный кусок кода, который исполняется при выполнении конкретного условия в программе. Чтобы обозначить этот кусок программы его выделяют с помощью проблелов и называют блоком.
Визуально блоки в Питоне видно сразу - это кусочки программы (кусочки кода), которые начинаются с одинакового отсутпа слева и идут один за другим (сверху вниз)

Без организации таких блоков невозможно, чтобы програма была ветвистой, то есть невозможно, чтобы Питон понимал какой кусок выполнить, если мы хотим, чтобы Питон его выполнил при наступления какого-то условия. Блоки используются при написании условных операторов и в циклах (циклы мы пройдем в следующей теме)
