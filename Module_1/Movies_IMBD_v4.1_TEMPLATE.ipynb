# Предобработка

```python
answers = {} # создадим словарь для ответов

# тут другие ваши предобработки колонок например:
# создаем новую калькулируемую колонку прибыли (прибыль = сбор - бюджет)
data['profit'] = data['revenue'] - data['budget']

#the time given in the dataset is in string format.
#So we need to change this in datetime format

data['release_date'] = pd.to_datetime(data['release_date'], format = '%m/%d/%Y')

""" Функция lst_from_col для формирования списка значений из столбца датасета, которое может содержать, 
    как единичное значение, так и список значений. На входе получает , датасет (dataset) индекс столбца (col_idx) 
    возвращает список """

def lst_from_col(dataset, col_idx):
    genre = []
    for row in dataset.itertuples():
        genre.extend(row[col_idx])
    return genre
```

# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['24'] = '5. Four By Two Productions'
# +
```

```python
# копируем датасет
data24 = data.copy()
# превращаем компании из строк с разделителем в список
data24.production_companies = data24.production_companies.str.split('|')
# меняем датасет для, создавая строк для каждой компании из списков
data24 = data24.explode('production_companies')
# вычисляем длину наименования фильма и добавляем в новое поле в датасет
data24['length'] = data24['original_title'].str.len()
# группируем студии по двум агрегатным функциям: сумма количества слов названия и число фильмов этой студии
group = data24.groupby('production_companies').agg({'length':'sum', 'original_title':'count'})
# вычисляем среднее значение длины названия (делим общее число слов на число фильмов выпущенных студией)
group['mean_length'] = group['length'] / group['original_title']
# выводим результат на экран без ограничения по числу строк   
with pd.option_context('display.max_rows', None, 'display.max_columns', 4):
    print(group[['mean_length']].sort_values(by='mean_length', ascending=False)) 
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
answers['25'] = '3. Midnight Picture Show'
# +
```

```python
data25 = data.copy()
# добавляем колонку со значением - количество слов в описании
data25['words_qty'] = data25['overview'].str.count(' ') +1 # Количество слов на один больше, чем пробелов в строке
# превращаем строку с о студиями с разделителем в список
data25['production_companies'] = data25['production_companies'].str.split('|')
# все студии из списка трансформируем в таблицу с единичной студией (без списка)
data25 = data25.explode('production_companies')
# группируем студии по двум агрегатным функциям: сумма количества слов описания и число фильмов этой студии
group = data25.groupby('production_companies').agg({'words_qty':'sum', 'original_title':'count'})
# вычисляем среднее значение длины описания (делим общее число слов на число фильмов выпущенных студией)
group['mean_words'] = group['words_qty'] / group['original_title']
# выводим результат на экран без ограничения по числу строк
with pd.option_context('display.max_rows', None, 'display.max_columns', 4):
    print(group[['mean_words']].sort_values(by='mean_words', ascending=False))
```

```python
data27 = data.copy()

# преобразуем всех актеров датасета в один единый список. Далее из списка будем формировать словарь с индексами актеров
cast_list = data27.cast.str.cat(sep='|').split('|')

# убираем дубликаты актеров из списка, прежде чем проиндексировать актеров
cast_list2 = pd.Series(cast_list).unique()

""" создаем словарь с ключом - ИМЯ АКТЕРА и индексом - ПОРЯДКОВЫЙ НОМЕР
    уникальные индексы актеров нужны, чтобы создать матрицу с колонками и столбцами поименнованными актерами.
    пересечение столбцов и колонок будет содержать число вхождений данной пары актеров в фильмах датасета """

cast_dict = {}
n = 0
for cast in cast_list2:
    if cast not in cast_dict.keys():
        cast_dict[cast] = n
        n += 1
        
# для удобства сохраняем данные датасета в поле с актерами в виде списка (убираем разделитель "|")
data27.cast = data27.cast.str.split('|')
        
# формируем нулевую матрицу-таблицу по числу уникальных актеров по горизонтали и по такому же числу по вертикали
len_of_squarematrix = len(cast_list2) # число строк и столбцов
square_matrix = [] # матрица квадратная: число строк = числу столбцоы
for y in range(len_of_squarematrix):
    emptylist = [0 for i in range(len_of_squarematrix)] # заполняем матрицу НУЛЯМИ
    square_matrix.append(emptylist)

# проходим сверху вниз по каждому полю датасета (по каждому фильму), вычисляем пары, и инкриминируем соответствующие ячейки матрицы
for row in data27.itertuples():
    list_of_names = list(combinations(row[5], 2))
    for a, b in list_of_names:
        key_x = cast_dict[b]
        key_y = cast_dict[a]
        square_matrix[key_y][key_x] += 1 # находим соответствующую ячейку, принадлежащую ПАРЕ и инкриминируем ее
        """ чтобы не обращать внимание на очередность следования
            имен в парах, зеркалируем значения в матрице """
        square_matrix[key_x][key_y] += 1 # зеркалируем значение

# поиск максимального числа встреченных пар актеров
max_n = 0
for y in range(len_of_squarematrix):
    for x in range(len_of_squarematrix):
        if square_matrix[y][x] > max_n:
            max_n = square_matrix[y][x]

# на случай, если максимумов несколько в матрице, наша задача выяснить их все и записать индексы в список
y_index = []
x_index = []
for y in range(len_of_squarematrix):
    for x in range(len_of_squarematrix):
        if x==y:
            continue # диагональ  игнорируем, т.к. связь "сам с собой" нам не нужна
        if square_matrix[y][x] == max_n:
            y_index.append(y)
            x_index.append(x)

# избавляемся от дубликатов появившихся в результате зеркалирования
x_index = pd.Series(x_index).unique()
y_index = pd.Series(y_index).unique()

# если есть хоть одна пара актеров, выводим результат
if max_n>0:
    for i in range(len(y_index)):
        cast1 = cast_list2[x_index[i]]
        cast2 = cast_list2[y_index[i]]
        print(f"Пара {cast1} и {cast2} чаще остальных пар актеров снимаются вместе")
```
